```python
import osmee2labels as oel
import os
import warnings
from shapely.errors import ShapelyDeprecationWarning
```

```python
## Used for training models
plg_bbox = (-73.963875,40.651940,-73.931259,40.672775)

## Used for testing models
park_slope_bbox = (-73.985505,40.664593,-73.972108,40.683359)
```

```python
## Only returns NRG bands for some reason (look into making code/model compatable with non-RGB bands)
philly_bbox = (-75.177985,39.935792,-75.144837,39.952390)
```

```python
## Get raster image as ee.Image object from bbox
img = oel.get_NAIP_raster_from_bbox(park_slope_bbox)
```

```python
## Get bbox of raster image
img_bbox = oel.get_Image_bbox(img)
```

```python
## Divide raster bbox into a 4x4 grid of 16 bboxes
sub_bboxs = oel.subdivide_bbox(img_bbox, tiles_x = 4, tiles_y = 4)
```

```python
## Names of directories to save raster and vectors to (don't have to exist yet)
rasters_dir = "./data/brooklyn/rasters/"
vectors_dir = "./data/brooklyn/vectors/"
```

```python
## Added to catch some of the DeprecationWarnings from dependancies
with warnings.catch_warnings():
    warnings.simplefilter('ignore', category=DeprecationWarning)
    warnings.simplefilter('ignore', category=ShapelyDeprecationWarning)
    
    ## Download rasters and vectors for each sub-bbox
    oel.export_subdivided_raster_vector(raster_image=img, bbox_list=sub_bboxs,
                                    raster_path_out=rasters_dir,
                                    vector_path_out=vectors_dir)
```

```python
rasters_paths = sorted([rasters_dir+file for file in os.listdir(rasters_dir)
                        if file.endswith(".tif")])

vectors_paths = sorted([vectors_dir+file for file in os.listdir(vectors_dir)
                        if file.endswith(".geojson")])

len(rasters_paths),len(vectors_paths)
```

```python
## Define directory to save masks to
masks_dir  = "./data/brooklyn/masks/"
```

```python
## Make masks from vector and raster files
oel.build_masks(source_raster_directory=rasters_dir,
            source_vector_directory=vectors_dir,
            mask_destination=masks_dir)
```

```python
masks_paths = sorted([masks_dir+file for file in os.listdir(masks_dir)
                       if file.endswith(".tif")])
len(masks_paths)
```

```python
## Define directories to save tiles to
tiled_images_dir = "./data/brooklyn/tiles/images/train/"
tiled_masks_dir  = "./data/brooklyn/tiles/masks/labels/"
```

```python
## Tile rasters and masks
oel.tile_tifs(raster_paths = rasters_paths, mask_paths = masks_paths,
          raster_tiles_out = tiled_images_dir,
          mask_tiles_out = tiled_masks_dir,
          tile_size = (256,256))
```

```python
tiled_img_paths = sorted([tiled_images_dir+file for file in os.listdir(tiled_images_dir)
                   if file.endswith(".tif")])

tiled_mask_paths = sorted([tiled_masks_dir+file for file in os.listdir(tiled_masks_dir)
                       if file.endswith(".tif")])

len(tiled_img_paths), len(tiled_mask_paths)
```

```python
import rasterio as rio
import matplotlib.pyplot as plt
import numpy as np


## plot images with corresponding masks to check results
def plot_img_with_mask(index, raster_img_paths = rasters_paths, mask_img_paths = masks_paths):
    tif_file = rio.open(raster_img_paths[index], mode = "r+").read().transpose(1,2,0)
    mask_file = rio.open(mask_img_paths[index], mode = "r+").read().transpose(1,2,0)
    plt.figure(figsize = (10,10))
    plt.imshow(tif_file)
    plt.imshow(mask_file, alpha =  0.75, cmap = "inferno")
```

```python
## Plot random raster tile with mask
plot_img_with_mask(np.random.choice(len(tiled_img_paths)),
                   raster_img_paths=tiled_img_paths,
                   mask_img_paths=tiled_mask_paths)
```

```python
## plot random larger subsection of the raster with mask (for a larger view of mask accuracy)
plot_img_with_mask(np.random.choice(len(rasters_paths)),
                   raster_img_paths=rasters_paths,
                   mask_img_paths=masks_paths)
```
